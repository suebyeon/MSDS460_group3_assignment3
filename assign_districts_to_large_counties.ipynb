```python
# if a district has a population that is larger than the target number for a district, 
# assign it at least one district

# target_pop for a district is the total state pop divided by the number of districts

# if a district is within 30% of a multiple of target_pop (over or under), assign it that many districts 

# e.g. (from assignment): Ohio target_pop = 737,466. Range for assigning a district is 737,466 +/- 30% = 

# 737,466 +/- 221240 = 516,226 - 958,706

# Cleveland, Ohio has a pop of 1,264,817. 

# 1,264,817 / 737,466 = 1 remainder 527351 --->  527351 is within 30% range so assign Cleveland 2 districts.
 
# Ohio's total population is 11,799,448. With 16 districts, we have a target population for each 
# district of 737,466. So we could assign two districts to Columbus (population 1,323,804) and 
# Cleveland (population 1,264,817) counties and one district to Cincinnati (population 830,639)county
```

```python
# calculate target pop for each district

num_districts = 9
total_pop = 6910840
target_pop = total_pop/num_districts

print(f'target_pop = {target_pop:.0f}')
```

```python
# lower bound for assigning a district to a county 

lower_bound = target_pop - (target_pop*0.3)
print(f'lower_bound = {lower_bound:.0f}')
```

```python
# upper bound for assigning a district to a county 

upper_bound = target_pop + (target_pop*0.3)
print(f'upper_bound = {upper_bound:.0f}')
```

```python
# divide the county population by the target pop for each district

df['pop_div_by_target'] = df['population']/target_pop
df.head(5)
```

```python
# get the reamainder of dividing county pop by target pop
# e.g. for Cleveland, Ohio the countypop % targetpop = remainder of 527351. 
# 527351 / 737,466 = 0.715

df['pop_div_target_remainder_percent'] = (df['population']%target_pop) / target_pop
df.head()
```

```python
df['num_districts_assigned'] = np.where(df['pop_div_target_remainder_percent'] > 0.7, 
                                        np.ceil(df['pop_div_by_target']), 
                                        np.floor(df['pop_div_by_target']))
df.head()
```
